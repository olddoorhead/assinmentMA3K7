```python
from math import floor
import numpy as np
import matplotlib.pyplot as plt
import math
import random
from statistics import mean 
import seaborn as sns
```

```python
def sequence_term(n):
    if n % 2 == 0: 
        k = n // 2
        return 1 + k * (k + 1)
    else: 
        k = floor(n / 2)
        return 1 + (k + 1) ** 2

# Generate the first 20 terms of the sequence
turning_pt = [sequence_term(n) for n in range(1, 51)]
print(turning_pt)
```

```python
def simulate_path(rolls):
    position = 1  # Start at position 1
    for _ in range(rolls):
        roll = random.randint(1, 6)
        position += roll
    return position

def num_turns(position):
    turns = 0  # Initialize turns
    for turning_point in turning_pt:
        if position >= turning_point:
            turns += 1  # Increment turns if the position is past the turning point
        else:
            break  # Stop the loop if the position is less than the turning point
    return turns

def probabilities(num_of_runs, rolls):
    sum = 0
    for _ in range(num_of_runs):
        position = simulate_path(rolls)
        if num_turns(position) >= rolls:
            sum += 1
    probability = sum / num_of_runs
    return probability

rolls = 2
print(probabilities(100, rolls))
rolls = 3
print(probabilities(100000, rolls))
rolls = 4
print(probabilities(100000, rolls))
rolls = 5
print(probabilities(100000, rolls))
```

```python
n_probability = []
for i in range(50):
    probability = probabilities(10000, i)
    n_probability.append(probability)

plt.plot(n_probability)
plt.xlabel('Number of Rolls')
plt.ylabel('Probability')
plt.title('Probability vs. Number of Rolls')
plt.show()
```

```python
for i in range(1,50):
    furtherest_step = 6*i+1
    if turning_pt[i+1] > furtherest_step:
        print(f'The number of rolls which P(n) starts become zero is:{i}')
        break
```

```python
specific_roll_count = 5

# Define the number of simulations and probability samples
num_of_runs = 1000
num_samples = 10000 

# Gather probability samples
probability_samples = [probabilities(num_of_runs, specific_roll_count) for _ in range(num_samples)]

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(probability_samples, kde=True, color='blue', binwidth=0.01)
plt.title(f'Probability Distribution for {specific_roll_count} Rolls')
plt.xlabel('Probability')
plt.ylabel('Frequency')
plt.show()
```
